```python
time_diff_list = []
a_list = []
iterate_num = 5
for i in range(len(parameter_array[:,1])):
    event_num= int(parameter_array[i,1])
    if event_num ==0:
        iterate_num +=1
    # handscan_time = hand_scan_t0_list[event_num]
    handscan_time = hand_scan_t0_list[i] # for both list has same length
    fitting_time = parameter_array[i,4]
    fitting_a = parameter_array[i,3]
    
    t_diff = fitting_time-handscan_time
    print(iterate_num,event_num,"handscan time", handscan_time, "fitting time", fitting_time,"t_diff", t_diff,"total i",i)
    if fitting_time>0:
        time_diff_list.append(t_diff)
        a_list.append(fitting_a)
    if t_diff >500:
        print("-600", event_num,fitting_time, handscan_time)
```

```python
fig, axes = plt.subplots()
# -700 to 700, every 10 ms
axes.hist(time_diff_list, bins=140, range=(-700,700))
axes.set_xlabel("t_diff value")
axes.set_ylabel("counts")
print(np.mean(time_diff_list))
print(np.std(time_diff_list))
print(len(time_diff_list))
# find the big discrepancy waveform and the camera
```
